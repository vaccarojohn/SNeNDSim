```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/dbscan_optimization_data.npz')

print("Successfully loaded data!")
```

```python
# Total energy deposited per event (in active volume)

# Define bins
bins = np.arange(-40, 21, 1)
plt.hist(data['data'][0,0,0], bins=bins, edgecolor='black')
plt.xlim(-40, 20)

# Axis labels
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Reconstructed Energy vs. GAMPixPy Energy (at epsilon=0.6, min_samples=4)')

plt.show()
```
