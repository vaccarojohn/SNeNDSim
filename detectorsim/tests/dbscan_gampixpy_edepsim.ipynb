```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read NumPy data
infile_dir = 'graph_data'
final_data = np.load(infile_dir + '/dbscan_final_data.npz')

print("Successfully loaded data!")
```

```python
# TPC Energy Reconstruction

# Define bins
bins = np.arange(-1, 1.02, 0.02)
plt.hist((final_data['data'][0,0,1] - final_data['data'][0,0,2])/final_data['data'][0,0,2], bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('Reconstructed Energy Difference (GAMPixPy vs. EdepSim)')
plt.ylabel('Frequency')
plt.title('TPC Energy Reconstruction (at epsilon=0.7, min_samples=3)')

plt.show()
```
