```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/dirt_segment_data.npz')
data2 = np.load('../../edepsim/dirt/' + infile_dir + '/dirt_segment_data.npz')

print("Successfully loaded data!")
```

```python
# Total energy deposited per event (in active volume)

# Define bins
bins = np.arange(0, 61, 1)
plt.hist(data['aenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 60)

# Axis labels
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Total Energy Deposited per Event (in Active Volume)')

plt.show()
```

```python
# Percent difference between reconstructed and true energy (in active volume)

# Define bins
bins = np.arange(-1, 1.025, 0.025)
plt.hist(data['aenergy'] / data2['aenergy'] - 1, bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('Percent Deviation from True Energy Deposition')
plt.ylabel('Frequency')
plt.title('Difference between Reconstructed and True Energy per Event (in Active Volume)')

plt.show()
```

```python
# Energy deposited in signal volume per event

# Define bins
bins = np.arange(0, 61, 1)
plt.hist(data['senergy'], bins=bins, edgecolor='black')
plt.xlim(0, 60)

# Axis labels
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Energy Deposited in Signal Volume per Event')

plt.show()
```

```python
# Percent difference between reconstructed and true energy (in signal volume)

# Define bins
bins = np.arange(-1, 1.025, 0.025)
plt.hist(data['senergy'] / data2['senergy'] - 1, bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('Percent Deviation from True Energy Deposition')
plt.ylabel('Frequency')
plt.title('Difference between Reconstructed and True Energy in Signal Volume per Event')

plt.show()
```

```python
# Energy deposited in fiducial volume per event

# Define bins
bins = np.arange(0, 61, 1)
plt.hist(data['fenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 60)

# Axis labels
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Energy Deposited in Fiducial Volume per Event')

plt.show()
```

```python
# Percent difference between reconstructed and true energy (in fiducial volume)

# Define bins
bins = np.arange(-1, 1.025, 0.025)
plt.hist(data['fenergy'] / data2['fenergy'] - 1, bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('Percent Deviation from True Energy Deposition')
plt.ylabel('Frequency')
plt.title('Difference between Reconstructed and True Energy in Fiducial Volume per Event')

plt.show()
```

```python
# Signal volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['psenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Share of Energy Deposited in Signal Volume per Event')

plt.show()
```

```python
# Difference between reconstructed and true fraction of energy deposited in signal volume

# Define bins
bins = np.arange(-1, 1.025, 0.025)
plt.hist(data['psenergy'] - data2['psenergy'], bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Difference between Reconstructed and True Share of Energy in Signal Volume per Event')

plt.show()
```

```python
# Fiducial volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['pfenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Share of Energy Deposited in Fiducial Volume per Event')

plt.show()
```

```python
# Difference between reconstructed and true fraction of energy deposited in fiducial volume

# Define bins
bins = np.arange(-1, 1.025, 0.025)
plt.hist(data['pfenergy'] - data2['pfenergy'], bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Difference between Reconstructed and True Share of Energy in Fiducial Volume per Event')

plt.show()
```
