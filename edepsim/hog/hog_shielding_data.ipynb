```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/hog_shielded_data.npz')
pad = 100000 - len(data['crttop'])

print("Successfully loaded data!")
```

```python
# Energy deposited in CRT volume per event
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2)

# Define bins
bins = np.arange(0, 1025, 25)

# CRT top
ax1.hist(np.pad(data['crttop']*1000, (0, pad)), bins=bins, edgecolor='black')
ax1.set_yscale('log')
ax1.set_xlim(0, 1000)
ax1.set_xlabel('Energy Deposited (keV)')
ax1.set_ylabel('Frequency')
ax1.set_title('Energy Deposited in CRT Top')

# CRT bottom
ax2.hist(np.pad(data['crtbottom']*1000, (0, pad)), bins=bins, edgecolor='black')
ax2.set_yscale('log')
ax2.set_xlim(0, 1000)
ax2.set_xlabel('Energy Deposited (keV)')
ax2.set_ylabel('Frequency')
ax2.set_title('Energy Deposited in CRT Bottom')

# CRT left
ax3.hist(np.pad(data['crtleft']*1000, (0, pad)), bins=bins, edgecolor='black')
ax3.set_yscale('log')
ax3.set_xlim(0, 1000)
ax3.set_xlabel('Energy Deposited (keV)')
ax3.set_ylabel('Frequency')
ax3.set_title('Energy Deposited in CRT Left')

# CRT right
ax4.hist(np.pad(data['crtright']*1000, (0, pad)), bins=bins, edgecolor='black')
ax4.set_yscale('log')
ax4.set_xlim(0, 1000)
ax4.set_xlabel('Energy Deposited (keV)')
ax4.set_ylabel('Frequency')
ax4.set_title('Energy Deposited in CRT Right')

# CRT front
ax5.hist(np.pad(data['crtfront']*1000, (0, pad)), bins=bins, edgecolor='black')
ax5.set_yscale('log')
ax5.set_xlim(0, 1000)
ax5.set_xlabel('Energy Deposited (keV)')
ax5.set_ylabel('Frequency')
ax5.set_title('Energy Deposited in CRT Front')

# CRT back
ax6.hist(np.pad(data['crtback']*1000, (0, pad)), bins=bins, edgecolor='black')
ax6.set_yscale('log')
ax6.set_xlim(0, 1000)
ax6.set_xlabel('Energy Deposited (keV)')
ax6.set_ylabel('Frequency')
ax6.set_title('Energy Deposited in CRT Back')

plt.tight_layout()
plt.show()
```

```python
# Energy deposited in CRT volume per event

# Define bins
bins = np.arange(0, 1010, 10)
plt.hist(np.pad(data['crt']*1000, (0, pad)), bins=bins, edgecolor='black')
plt.xlim(0, 1000)

# Axis labels
plt.yscale('log')
plt.xlabel('Energy Deposited (keV)')
plt.ylabel('Frequency')
plt.title('Energy Deposited in CRT')

plt.show()

print("Avg. Energy Deposited in CRT (total): " + str(np.sum(data['crt']*1000) / 100000) + "keV")
print("Avg. Energy Deposited in CRT (nonzero): " + str(np.sum(data['crt']*1000) / np.sum(data['crt'] != 0)) + "keV")
```
