```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/cosmics_vertex_data.npz')

print("Successfully loaded data!")
```

```python
# Energy deposited within 0.1cm of vertex

# Define bins
bins = np.arange(0, 10.25, 0.25)
plt.hist(data['e01cm'], bins=bins, edgecolor='black')
plt.xlim(0, 10)

# Axis labels
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Energy Deposited Within 0.1cm of Primary TPC Segment')

plt.show()
```

```python
# Share of energy deposited within 0.1cm of vertex

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['pe01cm'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Percentage of Energy Deposited Within 0.1cm of Primary TPC Segment')

plt.show()
```
