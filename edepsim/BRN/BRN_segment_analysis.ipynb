```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/BRN_segment_data.npz')

print("Successfully loaded data!")
```

```python
# Total energy deposited per event (in active volume)

# Define bins
bins = np.arange(0, 305, 5)
plt.hist(data['aenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 300)

# Axis labels
plt.yscale('log')
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Total Energy Deposited per Event (in Active Volume)')

plt.show()
```

```python
# Energy deposited in signal volume per event

# Define bins
bins = np.arange(0, 51, 1)
plt.hist(data['senergy'], bins=bins, edgecolor='black')
plt.xlim(0, 50)

# Axis labels
plt.yscale('log')
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Energy Deposited in Signal Volume per Event')

plt.show()
```

```python
# Energy deposited in fiducial volume per event

# Define bins
bins = np.arange(0, 51, 1)
plt.hist(data['fenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 50)

# Axis labels
plt.yscale('log')
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.title('Energy Deposited in Fiducial Volume per Event')

plt.show()
```

```python
# Primary neutron trajectory energy deposition plot

# Define bins
bins = np.arange(0, 1.55, 0.05)
plt.hist(data['prenergies'], bins=bins, edgecolor='black')
plt.xlim(0, 1.5)

# Axis labels
plt.yscale('log')
plt.xlabel('Energy Deposition (MeV)')
plt.ylabel('Frequency')
plt.title('Primary Neutron Energy Deposition (in Active Volume)')

plt.show()
```

```python
# Primary neutron trajectory length plot

# Define bins
bins = np.arange(0, 2.55, 0.05)
plt.hist(data['prlengths'], bins=bins, edgecolor='black')
plt.xlim(0, 2.5)

# Axis labels
plt.yscale('log')
plt.xlabel('Trajectory Length (cm)')
plt.ylabel('Frequency')
plt.title('Primary Neutron Trajectory Length (in Active Volume)')

plt.show()
```

```python
# Leading charged trajectory energy deposition plot

# Define bins
bins = np.arange(0, 46, 1)
plt.hist(data['lctes'], bins=bins, edgecolor='black')
plt.xlim(0, 45)

# Axis labels
plt.xlabel('Energy Deposition (MeV)')
plt.ylabel('Frequency')
plt.title('Leading Charged Trajectory (in Active Volume) Energy Deposition')

plt.show()
```

```python
# Leading charged trajectory length

# Define bins
bins = np.arange(0, 20.5, 0.5)
plt.hist(data['lctls'], bins=bins, edgecolor='black')
plt.xlim(0, 20)

# Axis labels
plt.xlabel('Trajectory Length (cm)')
plt.ylabel('Frequency')
plt.title('Leading Charged Trajectory Length (in Active Volume) per Event')

plt.show()
```

```python
# Longest trajectory plot

# Define bins
bins = np.arange(0, 20.5, 0.5)
plt.hist(data['lengths'], bins=bins, edgecolor='black')
plt.xlim(0, 20)

# Axis labels
plt.xlabel('Trajectory Length (cm)')
plt.ylabel('Frequency')
plt.title('Longest Energy Deposition Trajectory (in Active Volume) per Event')

plt.show()
```

```python
# Particles with highest energy depositions
plt.figure(figsize=(100, 20))
plt.bar([str(x) for x in data['eparticles']], data['ecounts'], edgecolor='black')

# Axis labels
plt.yscale('log')
plt.xlabel('PDG Code')
plt.ylabel('Frequency')
plt.title('Particles Resposible for Leading Charged Trajectory per Event')
plt.show()
```

```python
# Particles with longest trajectories
plt.figure(figsize=(100, 20))
plt.bar([str(x) for x in data['lparticles']], data['lcounts'], edgecolor='black')

# Axis labels
plt.yscale('log')
plt.xlabel('PDG Code')
plt.ylabel('Frequency')
plt.title('Particle Responsible for Longest Energy Deposition Trajectory')

plt.show()
```

```python
# Signal volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['psenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Share of Energy Deposited in Signal Volume per Event')

plt.show()
```

```python
# Fiducial volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['pfenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Share of Energy Deposited in Fiducial Volume per Event')

plt.show()
```

```python
# Number of HOG particles in TPC per event

# Define bins
bins = np.arange(0, 6, 1)
plt.hist(data['numhog'], bins=bins, edgecolor='black')
plt.xlim(0, 5)

# Axis labels
plt.yscale('log')
plt.xlabel('Number of HOGs')
plt.ylabel('Frequency')
plt.title('Number of HOGs in TPC per Event')

plt.show()

print("#HOG/event: " + str(np.sum(data['numhog'] != 0) / len(data['numhog'])))
```
