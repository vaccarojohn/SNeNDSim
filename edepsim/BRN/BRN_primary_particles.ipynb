```python
import h5py
import ROOT
import matplotlib.pyplot as plt
import numpy as np

def multiple_formatter(denominator=2, number=np.pi, latex='\pi'):
    def gcd(a, b):
        while b:
            a, b = b, a%b
        return a
    def _multiple_formatter(x, pos):
        den = denominator
        num = int(np.rint(den*x/number))
        com = gcd(num,den)
        (num,den) = (int(num/com),int(den/com))
        if den==1:
            if num==0:
                return r'$0$'
            if num==1:
                return r'$%s$'%latex
            elif num==-1:
                return r'$-%s$'%latex
            else:
                return r'$%s%s$'%(num,latex)
        else:
            if num==1:
                return r'$\frac{%s}{%s}$'%(latex,den)
            elif num==-1:
                return r'$\frac{-%s}{%s}$'%(latex,den)
            else:
                return r'$\frac{%s%s}{%s}$'%(num,latex,den)
    return _multiple_formatter
    
class Multiple:
    def __init__(self, denominator=2, number=np.pi, latex='\pi'):
        self.denominator = denominator
        self.number = number
        self.latex = latex
        
    def locator(self):
        return plt.MultipleLocator(self.number / self.denominator)
        
    def formatter(self):
        return plt.FuncFormatter(multiple_formatter(self.denominator, self.number, self.latex))
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/BRN_primary_particles.npz')

print("Successfully loaded data!")
```

```python
# Primary neutron energy plot

# Define bins
bins = np.arange(0, 405, 5)
plt.hist(data['nenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 400)

# Axis labels
plt.yscale('log')
plt.xlabel('Energy (MeV)')
plt.ylabel('Frequency')
plt.title('(Kinetic) Energy of Primary Neutrons')

plt.show()
```

```python
# Primary neutron direction plot ()

# Define bins
bins = np.arange(-1, 1.05, 0.05)
plt.hist(data['ndiry'], bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('cos()')
plt.ylabel('Frequency')
plt.title('Direction of Primary Neutrons (with respect to the positive y-axis)')

plt.show()
```

```python
# Primary neutron direction plot ()

# Define bins
bins = np.arange(0, 2.05*np.pi, 0.05*np.pi)
plt.hist(data['ndirz'], bins=bins, edgecolor='black')
plt.xlim(0, 2*np.pi)

ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 12))
ax.xaxis.set_major_formatter(plt.FuncFormatter(multiple_formatter()))

# Axis labels
plt.xlabel('')
plt.ylabel('Frequency')
plt.title('Direction of Primary Neutrons (with respect to the positive z-axis)')

plt.show()
```
