```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read NumPy data
infile_dir = 'graph_data'
signal_data = np.load('signal/' + infile_dir + '/signal_segment_data.npz')
cosmics_data = np.load('cosmics/' + infile_dir + '/cosmics_segment_data.npz')
BRN_data = np.load('BRN/' + infile_dir + '/BRN_segment_data.npz')

print("Successfully loaded data!")
```

```python
# Get number of signal, cosmic, and BRN events in the active volume

print("Number of Signal Events in TPC: " + str(len(signal_data['psenergy'])))
print("Number of Cosmic Events in TPC: " + str(len(cosmics_data['psenergy'])))
print("Number of BRN Events in TPC: " + str(len(BRN_data['psenergy'])))
```

```python
# Apply selection cut [SignalVolEFrac > 0.4]

n = 0
for psenergy in signal_data['psenergy']:
    if psenergy > 0.4:
        n += 1

print("Number of Signal Events with [SignalVolEFrac > 0.4]: " + str(n))

n = 0
for psenergy in cosmics_data['psenergy']:
    if psenergy > 0.4:
        n += 1

print("Number of Cosmic Events with [SignalVolEFrac > 0.4]: " + str(n))

n = 0
for psenergy in BRN_data['psenergy']:
    if psenergy > 0.4:
        n += 1

print("Number of BRN Events with [SignalVolEFrac > 0.4]: " + str(n))
```

```python
# Apply selection cut [SignalVolEFrac > 0.4 & SignalVolE > 5]

n = 0
for psenergy, senergy in np.column_stack((signal_data['psenergy'], signal_data['senergy'])):
    if psenergy > 0.4 and senergy > 5:
        n += 1

print("Number of Signal Events with [SignalVolEFrac > 0.4 & SignalVolE > 5]: " + str(n))

n = 0
for psenergy, senergy in np.column_stack((cosmics_data['psenergy'], cosmics_data['senergy'])):
    if psenergy > 0.4 and senergy > 5:
        n += 1

print("Number of Cosmic Events with [SignalVolEFrac > 0.4 & SignalVolE > 5]: " + str(n))

n = 0
for psenergy, senergy in np.column_stack((BRN_data['psenergy'], BRN_data['senergy'])):
    if psenergy > 0.4 and senergy > 5:
        n += 1

print("Number of BRN Events with [SignalVolEFrac > 0.4 & SignalVolE > 5]: " + str(n))
```

```python
# Apply selection cut [SignalVolEFrac > 0.4 & SignalVolE > 5 & MaxE > 5 & TotalE < 52 & TotalE > 10]

n = 0
for psenergy, senergy, lct, aenergy in np.column_stack((signal_data['psenergy'], signal_data['senergy'], signal_data['energies'], signal_data['aenergy'])):
    if psenergy > 0.4 and senergy > 5 and lct > 5 and aenergy < 52 and aenergy > 10:
        n += 1

print("Number of Signal Events with [SignalVolEFrac > 0.4 & SignalVolE > 5 & MaxE > 5 & TotalE < 52 & TotalE > 10]: " + str(n))

n = 0
for psenergy, senergy, lct, aenergy in np.column_stack((cosmics_data['psenergy'], cosmics_data['senergy'], cosmics_data['energies'], cosmics_data['aenergy'])):
    if psenergy > 0.4 and senergy > 5 and lct > 5 and aenergy < 52 and aenergy > 10:
        n += 1

print("Number of Cosmic Events with [SignalVolEFrac > 0.4 & SignalVolE > 5 & MaxE > 5 & TotalE < 52 & TotalE > 10]: " + str(n))

n = 0
for psenergy, senergy, lct, aenergy in np.column_stack((BRN_data['psenergy'], BRN_data['senergy'], BRN_data['energies'], BRN_data['aenergy'])):
    if psenergy > 0.4 and senergy > 5 and lct > 5 and aenergy < 52 and aenergy > 10:
        n += 1

print("Number of BRN Events with [SignalVolEFrac > 0.4 & SignalVolE > 5 & MaxE > 5 & TotalE < 52 & TotalE > 10]: " + str(n))
```
