```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/signal_segment_test_data.npz')

print("Successfully loaded data!")
```

```python
# Segment length plot

# Define bins
bins = np.arange(0, 3.1, 0.1)
plt.hist(data['lengths'], bins=bins, edgecolor='black')
plt.xlim(0, 3)

# Axis labels
plt.xlabel('Segment Length (cm)')
plt.ylabel('Frequency')
plt.title('Segment Length of Signal')

plt.show()
```
