```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/signal_segment_data.npz')

print("Successfully loaded data!")
```

```python
# Total energy deposited per event

# Define bins
bins = np.arange(0, 51, 1)
plt.hist(data['tenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 50)

# Axis labels
plt.xlabel('Total Energy Deposited')
plt.ylabel('Frequency')
plt.title('Total Energy Deposited per Event')

plt.show()
```

```python
# Leading charged trajectory energy deposition plot

# Define bins
bins = np.arange(0, 46, 1)
plt.hist(data['energies'], bins=bins, edgecolor='black')
plt.xlim(0, 45)

# Axis labels
plt.xlabel('Energy Deposition (MeV)')
plt.ylabel('Frequency')
plt.title('Leading Charged Trajectory Energy Deposition')

plt.show()
```

```python
# Longest trajectory plot

# Define bins
bins = np.arange(0, 20.5, 0.5)
plt.hist(data['lengths'], bins=bins, edgecolor='black')
plt.xlim(0, 20)

# Axis labels
plt.xlabel('Trajectory Length (cm)')
plt.ylabel('Frequency')
plt.title('Longest Energy Deposition Trajectory per Event')

plt.show()
```

```python
# Particles with highest energy depositions
plt.bar([str(x) for x in data['eparticles']], data['ecounts'], edgecolor='black')

# Axis labels
plt.yscale('log')
plt.xlabel('PDG Code')
plt.ylabel('Frequency')
plt.title('Particle Resposible for Highest Energy Deposition per Event')

plt.show()
```

```python
# Particles with longest trajectories
plt.bar([str(x) for x in data['lparticles']], data['lcounts'], edgecolor='black')

# Axis labels
plt.yscale('log')
plt.xlabel('PDG Code')
plt.ylabel('Frequency')
plt.title('Particle Responsible for Longest Energy Deposition Trajectory')

plt.show()
```

```python
# Fiducial volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['pfenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Share of Energy Deposited in Fiducial Volume per Event')

plt.show()
```

```python
# Signal volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['psenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Share of Energy Deposited in Signal Volume per Event')

plt.show()
```

```python
# Active volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['paenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.xlabel('Percentage of Energy Deposited')
plt.ylabel('Frequency')
plt.title('Share of Energy Deposited in Active Volume per Event')

plt.show()
```
