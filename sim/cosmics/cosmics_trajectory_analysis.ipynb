```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/cosmics_trajectory_data.npz')

print("Successfully loaded data!")
```

```python
# Energy change plot

# Define bins
bins = np.arange(0, 15250, 250)
plt.hist(data['echanges'], bins=bins, edgecolor='black')
plt.xlim(0, 15000)

# Axis labels
plt.yscale('log')
plt.xlabel('Energy Change (MeV)')
plt.ylabel('Frequency')
plt.title('Energy Change per Trajectory')

plt.show()
```

```python
# Trajectory length plot

# Define bins
bins = np.arange(0, 1625, 25)
plt.hist(data['lengths'], bins=bins, edgecolor='black')
plt.xlim(0, 1600)

# Axis labels
plt.yscale('log')
plt.xlabel('Trajectory Length (cm)')
plt.ylabel('Frequency')
plt.title('Trajectory Length')

plt.show()
```

```python
# Fiducial volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['pflengths'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.yscale('log')
plt.xlabel('Percentage of Trajectory Length')
plt.ylabel('Frequency')
plt.title('Percentage of Trajectory inside Fiducial Volume')

plt.show()
```

```python
# Signal volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['pslengths'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.yscale('log')
plt.xlabel('Percentage of Trajectory Length')
plt.ylabel('Frequency')
plt.title('Percentage of Trajectory inside Signal Volume')

plt.show()
```

```python
# Active volume ratio plot

# Define bins
bins = np.arange(0, 1.025, 0.025)
plt.hist(data['palengths'], bins=bins, edgecolor='black')
plt.xlim(0, 1)

# Axis labels
plt.yscale('log')
plt.xlabel('Percentage of Trajectory Length')
plt.ylabel('Frequency')
plt.title('Percentage of Trajectory inside Active Volume')

plt.show()
```
