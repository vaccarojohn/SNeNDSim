The following cells are graphs demonstrating the characteristics of primary electrons, photons, protons, and neutrons produced by neutrino interactions. The distribution of each primary particle's direction (i.e., its initial momentum) is expressed in two plots per particle. First, a plot is shown of cos(θ), where θ refers to the angle between the particle's momentum and the positive y-axis. The neutrino beam was modelled to propagate in the positive y-direction. Another plot is shown of ɸ, where ɸ refers to the angle between the particle's momentum and the positive z-axis on the xz-plane (i.e., rotated spherical coordinates).

```python
import h5py
import ROOT
import matplotlib.pyplot as plt
import numpy as np

def multiple_formatter(denominator=2, number=np.pi, latex='\pi'):
    def gcd(a, b):
        while b:
            a, b = b, a%b
        return a
    def _multiple_formatter(x, pos):
        den = denominator
        num = int(np.rint(den*x/number))
        com = gcd(num,den)
        (num,den) = (int(num/com),int(den/com))
        if den==1:
            if num==0:
                return r'$0$'
            if num==1:
                return r'$%s$'%latex
            elif num==-1:
                return r'$-%s$'%latex
            else:
                return r'$%s%s$'%(num,latex)
        else:
            if num==1:
                return r'$\frac{%s}{%s}$'%(latex,den)
            elif num==-1:
                return r'$\frac{-%s}{%s}$'%(latex,den)
            else:
                return r'$\frac{%s%s}{%s}$'%(num,latex,den)
    return _multiple_formatter
    
class Multiple:
    def __init__(self, denominator=2, number=np.pi, latex='\pi'):
        self.denominator = denominator
        self.number = number
        self.latex = latex
        
    def locator(self):
        return plt.MultipleLocator(self.number / self.denominator)
        
    def formatter(self):
        return plt.FuncFormatter(multiple_formatter(self.denominator, self.number, self.latex))
```

```python
# Read NumPy data
infile_dir = 'graph_data'
data = np.load(infile_dir + '/signal_primary_particles.npz')

print("Successfully loaded data!")
```

```python
# Primary electron multiplicity plot

# Define bins
bins = np.arange(1, 3, 1)
plt.hist(data['emult'], bins=bins, edgecolor='black')
plt.xlim(0, 3)

# Axis labels
plt.xlabel('Number of Primary Electrons')
plt.ylabel('Frequency')
plt.title('Number of Primary Electrons per Event')

plt.show()
```

```python
# Primary electron energy plot

# Define bins
bins = np.arange(0, 55, 1.25)
plt.hist(data['eenergy'], bins=bins, edgecolor='black')
plt.xlim(0, 55)

# Axis labels
plt.xlabel('Energy (MeV)')
plt.ylabel('Frequency')
plt.title('Energy of Primary Electrons')

plt.show()
```

```python
# Primary electron direction plot (θ)

# Define bins
bins = np.arange(-1, 1.05, 0.05)
plt.hist(data['ediry'], bins=bins, edgecolor='black')
plt.xlim(-1, 1)

# Axis labels
plt.xlabel('cos(θ)')
plt.ylabel('Frequency')
plt.title('Direction of Primary Electrons (with respect to the positive y-axis)')

plt.show()
```

```python
# Primary electron direction plot (ɸ)

# Define bins
bins = np.arange(0, 2.05*np.pi, 0.05*np.pi)
plt.hist(data['edirz'], bins=bins, edgecolor='black')
plt.xlim(0, 2*np.pi)

ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 12))
ax.xaxis.set_major_formatter(plt.FuncFormatter(multiple_formatter()))

# Axis labels
plt.xlabel('ɸ')
plt.ylabel('Frequency')
plt.title('Direction of Primary Electrons (with respect to the positive z-axis)')

plt.show()
```
